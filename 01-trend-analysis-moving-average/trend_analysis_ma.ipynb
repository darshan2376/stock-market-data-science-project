```python
# 📌 Step 1: Import libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8-darkgrid')

# 📌 Step 2: Download stock data
ticker = 'TCS.NS'
start_date = '2020-01-01'
end_date = '2024-12-31'

df = yf.download(ticker, start=start_date, end=end_date)
df.reset_index(inplace=True)

# 📌 Step 3: Calculate Moving Averages
df['MA20'] = df['Close'].rolling(window=20).mean()
df['MA50'] = df['Close'].rolling(window=50).mean()
df['MA200'] = df['Close'].rolling(window=200).mean()

# 📌 Step 4: Save dataset
df.to_csv('data/tcs_stock_data.csv', index=False)

# 📌 Step 5: Plot Closing Price with Moving Averages
plt.figure(figsize=(14, 7))
plt.plot(df['Date'], df['Close'], label='Close Price', color='blue')
plt.plot(df['Date'], df['MA20'], label='MA20', color='green', linestyle='--')
plt.plot(df['Date'], df['MA50'], label='MA50', color='orange', linestyle='--')
plt.plot(df['Date'], df['MA200'], label='MA200', color='red', linestyle='--')
plt.title('TCS Stock Price with Moving Averages')
plt.xlabel('Date')
plt.ylabel('Price (INR)')
plt.legend()
plt.savefig('images/trend_plot.png')  # Save plot
plt.show()
```
